# Exploratory Data Analysis on Categorical Variables
A deeper dive into waterfront and view

```python
import itertools
import numpy as np
import pandas as pd 
from numbers import Number
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OrdinalEncoder
%matplotlib inline 
```

```python
df = pd.read_csv("/Users/weatherford/Documents/Flatiron/Phase_2_Housing_Project/data/king_county_cleaned.csv")
water_df = df[['waterfront', "price"]]
```

```python
g = sns.FacetGrid(df, hue="waterfront", col="view_ord")

g.map(sns.scatterplot, "sqft_living", "price");
```

Subset of properties with the best views (view_ord = 4.0) has the greated concentration of waterfront property

```python
sns.barplot(data=df, x="waterfront", y="price", hue="view_ord");
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
sns.histplot(data=df, x="price", bins=100, hue = 'view_ord');
```
